```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from bokeh.plotting import figure, show, output_file
from bokeh.layouts import gridplot
from scipy import stats
```

```python
f = open("videodata-wgxmwm-x521b2.txt")

header = 0
lines = []

for l in f:
    if header == 0:
        header = l.split(" ")
        
        while '' in header:
            header.remove("")
        
        lines.append([x.strip('\n') for x in header])
        #lines = lines[lines != ]
    else:
        ln = []
        for d in l.split(" "):
            if( d == '' or d == "\n"):
                continue
            ln.append(d.strip('\n'))
        lines.append(ln)
f.close()

nf = open("videodata.csv", "w")

for vector in lines:
    strl = ""
    for elem in vector:
        strl = strl + str(elem) + ","
    strl = strl[:-1] + "\n"
    nf.write(strl)

nf.close()
```
